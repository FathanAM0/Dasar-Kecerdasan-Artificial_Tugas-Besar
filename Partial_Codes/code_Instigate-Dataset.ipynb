```python
import json
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf
```

```python
SAMPLE_RATE = 48000
SEGMENT_DURATION = int(1 * SAMPLE_RATE)
OVERLAP_DURATION = int(0.5 * SAMPLE_RATE)
```

```python
dir_data_nad = r"C:\Users\Lulay\Documents\GitHub\Dasar-Kecerdasan-Artificial_Tugas-Besar\Dataset\noise-audio-data"
dir_data_xc = r"C:\Users\Lulay\Documents\GitHub\Dasar-Kecerdasan-Artificial_Tugas-Besar\Dataset\xeno-canto"
dir_label_nad = r"C:\Users\Lulay\Documents\GitHub\Dasar-Kecerdasan-Artificial_Tugas-Besar\Dataset\nad_label.json"
dir_label_xc = r"C:\Users\Lulay\Documents\GitHub\Dasar-Kecerdasan-Artificial_Tugas-Besar\Dataset\xc_label.json"
```

```python
with open(dir_label_nad) as json_file:
    data_nad = json.load(json_file)

with open(dir_label_xc) as json_file:
    data_xc = json.load(json_file)
```

```python
def load_file_audio(path):
    audio, sr = sf.read(path)
    return np.array(audio), sr
```

```python
def plot_audio_with_loud_segments(audio, sr, list_result):
    time_axis = np.linspace(0, len(audio) / sr, len(audio))

    plt.figure(figsize=(14, 4))
    plt.plot(time_axis, audio, label="Amplitude")
    plt.title("Audio waveform with loud segments marked")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")

    # Corrected red span drawing
    for i in enumerate(list_result):
        start_time = i[1][0]
        end_time = i[1][1]
        plt.axvspan(start_time, end_time, color='red', alpha=0.3)

    plt.tight_layout()
    plt.legend()
    plt.show()
```

```python
for file_path in data_xc:
    file_full_path = dir_data_xc + "/" + file_path
    audio, sr = load_file_audio(file_full_path)
    plot_audio_with_loud_segments(audio, sr, data_xc[file_path])
```

```python
for file_path in data_nad:
    file_full_path = dir_data_nad + "/" + file_path
    audio, sr = load_file_audio(file_full_path)
    plot_audio_with_loud_segments(audio, sr, data_nad[file_path])
```
